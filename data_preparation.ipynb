```python
fisu['d.o.b.'] = fisu['d.o.b.'].dt.year
juniors['d.o.b.'] = juniors['d.o.b.'].dt.year
```

Calculating the ages of athletes:

```python
fisu['Age'] = COMPETITIONS_YEAR -  fisu['d.o.b.']
juniors['Age'] = COMPETITIONS_YEAR -  juniors['d.o.b.']
open['Age'] = COMPETITIONS_YEAR - open['d.o.b.']
masters['Age'] = COMPETITIONS_YEAR -  masters['d.o.b.']
```
