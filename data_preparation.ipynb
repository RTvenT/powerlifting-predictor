```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
```

```python
def clear_table(file_path):
    wb = openpyxl.load_workbook(file_path)

    for sheetname in wb.sheetnames:

        ws = wb[sheetname]
        print(f"Processing: {sheetname}")

        min_row = 2 if sheetname=='Table 1' else 1
        


        for row in ws.iter_rows(min_row=min_row):
            for cell in row:
                if cell.column_letter not in ['A', 'B']:
                    if cell.font.strike or isinstance(cell.value, str):
                        cell.value = None

    wb.save(file_path[:-5] + '_clear.xlsx')
```

```python
clear_table('./data/FISUWorldMensUniversity2024.xlsx')
clear_table('./data/WorldJuniorsMensClassic2024.xlsx')
clear_table('./data/WorldMastersMensClassic2024.xlsx')
clear_table('./data/WorldOpenMensClassic2024.xlsx')
```

```python
fisu = pd.read_excel("./data/FISUWorldMensUniversity2024_clear.xlsx")
juniors = pd.read_excel("./data/WorldJuniorsMensClassic2024_clear.xlsx")
open = pd.read_excel("./data/WorldOpenMensClassic2024_clear.xlsx")
masters = pd.read_excel("./data/WorldMastersMensClassic2024_clear.xlsx")
```
